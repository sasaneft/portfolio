```python
# Import requests package:
import requests
```

```python
# Set URL as url:
url = 'https://www.ec.europa.eu/agrifood/api/rawMilk/prices?'
```

```python
# Create parameter dictionary:
parameters = {
    'beginDate': '01/01/2021',
    'endDate': '31/12/2023'}
```

```python
# Send get request and save in variable r:
r = requests.get(url, parameters)

# Print result:
print(r)
```

```python
# Print cookies:
r.cookies
```

```python
# Apply JSON decoder and save in prices_milk_prelim:
prices_milk_prelim = r.json()

# Print prices_milk_prelim:
prices_milk_prelim
```

```python
# We do not need the variables "month", "monthName", and "year",
# especially as we do not know their exact relationships with beginDate and endDate:
prices_milk = prices_milk.drop(['month','monthName','year'], axis=1)

# We do not need the unit variable, as we know that prices are measured in €/100kg:
prices_milk = prices_milk.drop(['unit'], axis=1)

# Rename variable names to standard format and to names we can easily work with:
prices_milk = prices_milk.rename(columns={'memberStateCode': 'country_code'})
prices_milk = prices_milk.rename(columns={'memberStateName': 'country'})
prices_milk = prices_milk.rename(columns={'beginDate': 'begin_date'})
prices_milk = prices_milk.rename(columns={'endDate': 'end_date'})
prices_milk = prices_milk.rename(columns={'product': 'product_type'})

# Show variables again:
prices_milk.info()
```

```python
# Remove the € symbol in the price values:
prices_milk['price'] = prices_milk['price'].str.replace('€', '')

# Check results:
prices_milk
```

```python
# Transform begin_date and end_date to datetime format:
prices_milk['begin_date'] = pd.to_datetime(prices_milk['begin_date'], format='%d/%m/%Y')
prices_milk['end_date'] = pd.to_datetime(prices_milk['end_date'], format='%d/%m/%Y')

# Transform price to float:
prices_milk['price'] = prices_milk['price'].astype(float)

# Check results:
prices_milk.info()
```

```python
# Transform prices from the the unit €/100kg to the unit €/kg:
prices_milk['price'] = prices_milk['price'] / 100

# Check results:
prices_milk
```

```python
# We do not need the aggregated "EU-UK" prices in the dataframe:
prices_milk = prices_milk[(prices_milk['country_code'] != "EU-UK")]

# Check result:
prices_milk.country_code.unique()
```

```python
# We give simpler and more meaningful names to the product types (organic vs. conventional):
prices_milk['product_type'] = prices_milk['product_type'].str.replace('Organic raw milk', 'Organic')
prices_milk['product_type'] = prices_milk['product_type'].str.replace('Raw milk', 'Conventional')

# Check result:
prices_milk.product_type.unique()
```

```python
# Now we calculate price premia (i.e., relative price differences) for organic vs. conventional milk.
```

```python
# In order to calculate price premia, we need separate variables for
# the prices of organic and conventional milk, so that we can relate the prices to each other.
```

```python
# Create dataframe specifically for organic milk:
prices_milk_organic = prices_milk[prices_milk['product_type'] == "Organic"]

# Rename price variable to specify the product type:
prices_milk_organic = prices_milk_organic.rename(columns={'price': 'price_organic'})

# Show dataframe:
prices_milk_organic
```

```python
# Create dataframe specifically for conventional milk:
prices_milk_convent = prices_milk[prices_milk['product_type'] == "Conventional"]

# Rename price variable to specify the product type:
prices_milk_convent = prices_milk_convent.rename(columns={'price': 'price_convent'})

# Show dataframe:
prices_milk_convent
```

```python
# Now we step-wise merge the prices for organic and conventional milk (for each date and country_code) to the prices_milk dataframe:
prices_milk = pd.merge(prices_milk, prices_milk_organic[['begin_date','end_date','country_code','price_organic']], on=['begin_date','end_date','country_code'], how='left')
prices_milk = pd.merge(prices_milk, prices_milk_convent[['begin_date','end_date','country_code','price_convent']], on=['begin_date','end_date','country_code'], how='left')

# Show dataframe:
prices_milk
```

```python
# Price premium for organic vs. conventional milk in percent:
prices_milk['price_organic_p'] = ( ( prices_milk['price_organic'] - prices_milk['price_convent'] ) / prices_milk['price_convent'] ) * 100

# Show results:
prices_milk
```

```python
# The variables "price_organic" and "price_convent" are not needed anymore:
prices_milk = prices_milk.drop(['price_organic','price_convent'], axis=1)
```

```python
# Now we upload the dataframe to the database on the server.
```

```python
# Import sql_functions.py because we need some functions from that module:
import sql_functions as sqlf

# We need to restart the kernel and rerun at this point if we changed the module since we first imported it.
```

```python
# Create a variable called engine using the get_engine function:
engine = sqlf.get_engine()
```

```python
# We set the schema to our project name:
schema = 'capstone_organicfood'

# We set the table_name variable to the name of the dataframe:
table_name = 'prices_milk'
```

```python
# We need psycopg2 for raising possible error message:
import psycopg2
```

```python
# Write records stored in the dataframe to SQL database:
if engine!=None:
    try:
        prices_milk.to_sql(name=table_name, # name of SQL table variable
                        con=engine, # engine or connection
                        schema=schema, # our class schema variable
                        if_exists='replace', # Drop the table before inserting new values
                        index=False, # Write DataFrame index as a column
                        chunksize=5000, # Specify the number of rows in each batch to be written at a time
                        method='multi') # Pass multiple values in a single INSERT clause
        print(f"The {table_name} table was imported successfully.")
    # Error handling
    except (Exception, psycopg2.DatabaseError) as error:
        print(error)
        engine = None
else:
    print('No engine')
```

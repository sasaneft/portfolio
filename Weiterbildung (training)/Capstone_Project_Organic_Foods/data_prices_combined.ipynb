```python
# We first load the prices_eggs and prices_milk tables from the database into dataframes.
```

```python
# Import sql_functions.py because we need some functions from that module:
import sql_functions as sqlf

# We need to restart the kernel and rerun at this point if we changed the module since we first imported it.
```

```python
# Define the schema of our database:
schema = 'capstone_organicfood'
```

```python
# Load prices_eggs table into dataframe:
prices_eggs = sqlf.get_dataframe(f'SELECT * FROM {schema}.prices_eggs;')
```

```python
# Show dataframe:
prices_eggs
```

```python
# Show variables:
prices_eggs.info()
```

```python
# Load prices_milk table into dataframe:
prices_milk = sqlf.get_dataframe(f'SELECT * FROM {schema}.prices_milk;')
```

```python
# Show dataframe:
prices_milk
```

```python
# Show variables:
prices_milk.info()
```

```python
# Generate variable to specify the product (eggs) in prices_eggs dataframe:
prices_eggs['product'] = 'Eggs'

# Show dataframe:
prices_eggs
```

```python
# Generate variable to specify the product (raw milk) in prices_milk dataframe:
prices_milk['product'] = 'Raw milk'

# Show dataframe:
prices_milk
```

```python
# Import pandas package:
import pandas as pd
```

```python
# We combine the two dataframes into a new dataframe, using concat:
prices = pd.concat([prices_eggs, prices_milk])

# Show combined dataframe:
prices
```

```python
# Now we upload the dataframe to the database on the server.
```

```python
# Create a variable called engine using the get_engine function:
engine = sqlf.get_engine()
```

```python
# We set table_name to the name of the dataframe:
table_name = 'prices'
```

```python
# We need psycopg2 for raising possible error message:
import psycopg2
```

```python
# Write records stored in dataframe to SQL database:
if engine!=None:
    try:
        prices.to_sql(name=table_name, # name of SQL table variable
                        con=engine, # engine or connection
                        schema=schema, # our class schema variable
                        if_exists='replace', # Drop the table before inserting new values
                        index=False, # Write DataFrame index as a column
                        chunksize=5000, # Specify the number of rows in each batch to be written at a time
                        method='multi') # Pass multiple values in a single INSERT clause
        print(f"The {table_name} table was imported successfully.")
    # Error handling
    except (Exception, psycopg2.DatabaseError) as error:
        print(error)
        engine = None
else:
    print('No engine')
```

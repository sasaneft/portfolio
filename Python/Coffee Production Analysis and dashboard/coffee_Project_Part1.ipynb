# Coffee Consumption Vs. Prices

Compare coffee prices with coffee consumption in the following countries. 

Is there a relationship between coffee consumption and the price these countries pay?

Plot a dual axis bar chart of price vs consumption totals for these countries.

```python
consumption = pd.read_csv("./imports.csv").set_index("imports").mean(axis=1)
consumption.name="imports"
consumption.head()
```

```python
prices = pd.read_csv("./retail-prices.csv")

prices.head()
```

```python
prices = pd.read_csv("./retail-prices.csv").set_index("retail_prices").mean(axis=1)
prices.head()
```

```python
# merging twoe data frames together
price_cons=(pd.DataFrame(prices)
 .merge(consumption, 
        left_on=prices.index, 
        right_on=consumption.index, 
        how="inner")
 .sort_values(["imports"], ascending=False))

price_cons.columns=["Country", "Price", "Imports"]

#remove indentations to make labels more clear
price_cons["Country"] = price_cons["Country"].str.strip(" ")
price_cons
```

```python
fig, ax=plt.subplots(figsize=(14,10))
width = 0.35
x =np.arange(0, len(price_cons["Country"])) # for x-axis tick labels

bar1 = ax.bar(x-width/2, price_cons["Imports"], width=width, color="#6495ED")
ax.set_title("Price Paid vs. Consumprion of Select Importing Nations", color="grey", fontsize=18, fontweight="bold")
ax.set_ylabel("Consumption (1000 60Kg bags)", color="grey", fontsize=12)

ax2= ax.twinx()

bar2= ax2.bar(x+width/2, price_cons["Price"], width=width, color="#40E0D0")
ax2.set_ylabel("Average Price Paid per Bag (USD)", fontsize=12, color="grey")

plt.xticks(x, price_cons["Country"], fontsize=7)

plt.legend([bar1, bar2],["Consumption", "Price"] );
```
